### 4.3.1 Exploring Numerical Columns : Univariate Analysis

```python
# Defining a Helper Function

def uni_num_analysis(data, column_name, bins=10,show_lables=False):
    """
    Plots Histogram & Box Plots for the given dataset and column.
    
    data : dataset
    column_name : single column from the dataset
    
    """
    
    dataset = data
    value = str(column_name)
    
    # Subplots
    figure,axis = plt.subplots(1,2, figsize=(15,6))
    
    #Plot 1 - Histogram
    axis[0] = dataset[value].plot.hist(ax=axis[0],bins=bins)
    axis[0].set_title(f"{column_name} Frequency Plot")
    # set the bar label
    if show_lables == True:
        for c in axis[0].containers:
            axis[0].bar_label(c, label_type='edge')
    
    # Plot 2 Box plots
    axis[1] = sns.boxplot(x=dataset[value],ax=axis[1])
    axis[1].set_title(f"{column_name} Box & Whisker Plot")
    
    figure.tight_layout()
    plt.show()
```

### Quanity Column

```python
uni_num_analysis(clean_data,'Quantity',bins=13)
```

### Sales Column

```python
uni_num_analysis(clean_data,'Sales',bins=50)
```

### Discount Column

```python
uni_num_analysis(clean_data,'Discount',bins=8,show_lables=True)
```

### Profit Column

### 4.3.5 Exploring Numerical & Categorical Data Together : Multivariate Analysis

To see the distribution among the different categorical and numerical variables together.

```python
# Defining a Helper Function

def four_plot(data,name='Input',show_lables=False,rotation=0):
    # Fix The Subplot Size and Inputs
    figure, axis = plt.subplots(2,2, figsize=(15,10))
    
    # For First Plot
    axis[0,0] = sns.barplot(x=data.index,y=data['Sales'], ax=axis[0,0])
    axis[0,0].set_title(f"{name} vs Total Sales")
    axis[0,0].set_xticklabels(labels=data.index, rotation=rotation)
    # set the bar label
    if show_lables == True:
        for c in axis[0,0].containers:
            axis[0,0].bar_label(c, label_type='center')
    
    # For Second Plot
    axis[0,1] = sns.barplot(x=data.index,y=data['Quantity'], ax=axis[0,1])
    axis[0,1].set_title(f"{name} vs Total Quantity")
    axis[0,1].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[0,1].containers:
            axis[0,1].bar_label(c, label_type='center')
    
    # For Third Plot
    axis[1,0] = sns.barplot(x=data.index,y=data['Discount'], ax=axis[1,0])
    axis[1,0].set_title(f"{name} vs Total Discount")
    axis[1,0].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[1,0].containers:
            axis[1,0].bar_label(c, label_type='center')
    
    # For Fourth Plot
    axis[1,1] = sns.barplot(x=data.index,y=data['Profit'], ax=axis[1,1])
    axis[1,1].set_title(f"{name} vs Total Profit")
    axis[1,1].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[1,1].containers:
            axis[1,1].bar_label(c, label_type='center')
    
    figure.tight_layout()
    plt.show()
```

### Ship Mode

### Profit/Sales Ratio with respect to Category and Subcategory 

```python
# Table Formation
cat_and_sub_profit = clean_data.groupby(['Category','Sub-Category']).sum()[['Profit']].rename(columns = {'Profit':'Ratio'})
cat_and_sub_sales = clean_data.groupby(['Category','Sub-Category']).sum()[['Sales']].rename(columns = {'Sales':'Ratio'})
table = ((cat_and_sub_profit/cat_and_sub_sales)*100).sort_values(by=['Ratio'],ascending=True)

# Table Plot
fig,ax = plt.subplots()
ax = table.plot.barh(ax=ax);
ax.set_title("Profit/Sales Ratio")
for c in ax.containers:
    ax.bar_label(c, label_type='edge')
fig.tight_layout()
plt.show()
```

**Note :** We can confirm that the profit vs sales ratio for three subcategories are in negetive as seen in the previous graphs as they are incurring loss to the sales.

### For States & Cities

```python
# Defining a Helper Function

def top_25_plot(data,name='Input',show_lables=False,rotation=0):
    # Fix The Subplot Size and Inputs
    figure, axis = plt.subplots(2,2, figsize=(15,10))
    
    # For First Plot
    data = data.sort_values(by=['Sales'],ascending=False)[0:25]
    axis[0,0] = sns.barplot(x=data.index,y=data['Sales'], ax=axis[0,0])
    axis[0,0].set_title(f"{name} vs Total Sales")
    axis[0,0].set_xticklabels(labels=data.index, rotation=rotation)
    # set the bar label
    if show_lables == True:
        for c in axis[0,0].containers:
            axis[0,0].bar_label(c, label_type='center')
    
    # For Second Plot
    data = data.sort_values(by=['Quantity'],ascending=False)[0:25]
    axis[0,1] = sns.barplot(x=data.index,y=data['Quantity'], ax=axis[0,1])
    axis[0,1].set_title(f"{name} vs Total Quantity")
    axis[0,1].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[0,1].containers:
            axis[0,1].bar_label(c, label_type='center')
    
    # For Third Plot
    data = data.sort_values(by=['Discount'],ascending=False)[0:25]
    axis[1,0] = sns.barplot(x=data.index,y=data['Discount'], ax=axis[1,0])
    axis[1,0].set_title(f"{name} vs Total Discount")
    axis[1,0].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[1,0].containers:
            axis[1,0].bar_label(c, label_type='center')
    
    # For Fourth Plot
    data = data.sort_values(by=['Profit'],ascending=False)[0:25]
    axis[1,1] = sns.barplot(x=data.index,y=data['Profit'], ax=axis[1,1])
    axis[1,1].set_title(f"{name} vs Total Profit")
    axis[1,1].set_xticklabels(labels=data.index, rotation=rotation)
    if show_lables == True:
        for c in axis[1,1].containers:
            axis[1,1].bar_label(c, label_type='center')
    
    figure.tight_layout()
    plt.show()
```

### Top States

```python
state_data = clean_data.groupby(['State']).sum()[['Sales', 'Quantity', 'Discount', 'Profit']]
top_25_plot(state_data,name="State",show_lables=False, rotation=90)
```

### States with Profit/Sales Ratio

```python
# Table Formation
state_profit = clean_data.groupby(['State']).sum()[['Profit']].rename(columns = {'Profit':'Ratio'})
state_sales = clean_data.groupby(['State']).sum()[['Sales']].rename(columns = {'Sales':'Ratio'})
table1 = ((state_profit/state_sales)*100).sort_values(by=['Ratio'],ascending=False)

# Table Plot
fig,ax = plt.subplots(figsize=(15,7))
ax = table1.plot.bar(ax=ax)
ax.set_title("State wise Profit/Sales Ratio")
fig.tight_layout()
plt.show()
```

### Top Cities

```python
city_data = clean_data.groupby(['City']).sum()[['Sales', 'Quantity', 'Discount', 'Profit']]
top_25_plot(city_data,name="City",show_lables=False, rotation=90)
```

### Cities with Profit/Sales Ratio

```python
# Table Formation
city_profit = clean_data.groupby(['City']).sum()[['Profit']].rename(columns = {'Profit':'Ratio'})
city_sales = clean_data.groupby(['City']).sum()[['Sales']].rename(columns = {'Sales':'Ratio'})
table1 = ((city_profit/city_sales)*100).sort_values(by=['Ratio'],ascending=False)[0:25]
table2 = ((city_profit/city_sales)*100).sort_values(by=['Ratio'],ascending=True)[0:25]

# Table Subplots
fig,ax = plt.subplots(1,2,figsize=(15,7))
ax[0] = table1.plot.bar(ax=ax[0])
ax[0].set_title("Top 25 Citites by Profit/Sales Ratio")

ax[1] = table2.plot.bar(ax=ax[1])
ax[1].set_title("Bottom 25 Cities by Profit/Sales Ratio")

fig.tight_layout()
plt.show()
```
